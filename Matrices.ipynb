# Solving a Matrix Equation Ax=b

## Using Solve

```python

from numpy import *
from numpy.linalg import *
A=([[1,2,3],[22,32,42],[55,66,100]])
b=([1,2,3])
x=solve(A,b)
print('x = ',x)
y=dot(A,x)-b
print('y = ',y)
```

## By finding inv of A

```python
x=dot(inv(A),b)
print('x =',x)
y=dot(A,x)-b
print('y =',y)
```

# Eigen Value Problem $$I\omega=\lambda \omega$$

```python
I=array([[2/3,-1/4],[-1/4,2/3]])
Es,evectors=eig(I)
print('Eigen Value = '+str(Es))
print('')
print('Eigen vector matrix = \n'+str(evectors))
print('')
# Just Confirming below
Vec=array([evectors[0,0],evectors[1,0]])
print('Eigen Vectors ='+str(Vec)) #Only unique values
print('')
LHS=dot(I,Vec)
RHS=dot(Es[0],Vec)
Confirmation=LHS-RHS
print('Confirmation = '+str(Confirmation))
```

```python
A=array([-2,2,-3,2,1,-6,-1,-2,0])
A=A.reshape(3,3)
Es,evectors=eig(A)
print('Es = \n'+str(Es))
print('   ')
print('evectors matrix = \n'+str(evectors))
print('   ')
x=dot(Es[1],evectors[0:3,2])
print('x1 = \n'+str(x))
print('   ')
analytical=array([-1,-2,1]).reshape(3,1)/sqrt(6)
print('Analytical =\n '+str(analytical))
print('   ')
LHS=dot(A,analytical)
print('LHS = \n'+str(LHS))
print('   ')
RHS=dot(Es[1],analytical)
print('RHS = \n'+str(RHS))
print('   ')
print('LHS = RHS proves that the corresponding eigen vector being proportional to eigenvalue =5 ')
print('   ')
analytical=array([-2,1,0]).reshape(3,1)/sqrt(5)
LHS=dot(A,analytical)
print('LHS = \n'+str(LHS))
print('   ')
RHS=dot(Es[0],analytical)
print('RHS = \n'+str(RHS))
print('   ')
print('LHS = RHS proves that the corresponding eigen vector being proportional to eigenvalue = -3 ')
print('   ')
analytical=array([3,0,1]).reshape(3,1)/sqrt(10)
LHS=dot(A,analytical)
print('LHS = \n'+str(LHS))
print('   ')
RHS=dot(Es[2],analytical)
print('RHS = \n'+str(RHS))
print('   ')
print('LHS = RHS proves that the corresponding eigen vector being proportional to eigenvalue = -3 ')
print('   ')
```

### Physical system with N couples equations and N unknowns (Hilbert Matrices)

```python
A=array([])
for i in range(1,101):
    for j in range(1,101):
        a=1/(i+j-1)
        A=append(A,a)
A=A.reshape((100,100))
print('A =\n'+str(A))
print('  ')
B=A[0:100,0].reshape(100,1)
print('B = \n'+str(B))
print('  ')
y=solve(A,B)
print('y = \n'+str(y))
```
