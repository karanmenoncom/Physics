```python
ima=Image.open('/home/karanmenoncom/Python/newwhiteimage2.jpg')
np_im=array(ima)
np_im[999][999]=[0,0,0]
n=5
k=3
for i in range(n):
    print('interation',i)
    for x in range(1000):
        for y in range(1000):
            if np_im[x][y][0]==0:
                for u in range(1,5):
                    np_im[x-(u*int(k**(i+1)))][y]=[0,0,0]
                    if u==1 or u == 3:
                        np_im[x-(u*int(k**(i+1)))][y-(1*int(k**(i+1)))]=[0,0,0]
                    if u==2 :
                        np_im[x-(u*int(k**(i+1)))][y-(2*int(k**(i+1)))]=[0,0,0]
                    
            
new_im=Image.fromarray(np_im)
new_im.save('/home/karanmenoncom/Python/triangle5.jpg')
```

```python
from numpy import *
from pylab import *
import random as random
```

```python
%matplotlib inline
vertices=array([(0,0),(0.5,1),(1,0)])
pointsX=array([])
pointsY=([])
n=10000
x,y=random.choice(vertices)
for i in range(n):
    vx,vy=random.choice(vertices)
    x=(vx+x)/2
    y=(vy+y)/2
    pointsX=append(pointsX,x)
    pointsY=append(pointsY,y)
scatter(pointsX,pointsY,s=0.5)
```
