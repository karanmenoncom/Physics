```python
#X=arange(-10,10,0.01)
#Y=arange(-10,10,0.01)
T=arange(0,200,0.01)
def fx():
    return vx
def fy():
    return vy
def fvx():
    return -G*M*x/((x*x+y*y)**(3/2))
def fvy():
    return -G*M*y/((x*x+y*y)**(3/2))
x,y,vx,vy=0.5,0,0.01,1.62
X=([])
Y=([])
VX=([]);VY=([])
G=1 ;M=1
dt=0.01
for t in T:
    x=x+dt*(fx())
    y=y+dt*(fy())
    vx+=dt*(fvx())
    vy+=dt*(fvy())
    X=append(X,x)
    Y=append(Y,y)
    VX=append(VX,vx)
    VY=append(VY,vy)
Z=zeros(len(X))
fig=p.figure()
ax=Axes3D(fig)
ax.plot(X,Y,Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
#ax.view_init(20,20)
p.show()
```

## Logistic Map

```python
l=3
x=0.4
eps=10**(-2)
X=([])
Y=([])
scatter(x,x,s=15,color='black')
for t in range(0,100):
    new_x=l*x*(1-x)
    plot([x,x],[x,new_x],color='black')
    plot([x,new_x],[new_x,new_x],color='black')
    #scatter(x,new_x)
    a=x
    x=new_x
if a-x<=eps:
    print('Value converges to',x)
else:
    print('Chaotic')
X=arange(0,1.01,0.01)
Y=l*X*(1-X)
plot(X,Y,color='grey',linestyle='dotted',linewidth=1.5)
plot(X,X,color='grey',linestyle='dashed',linewidth=2)
xlabel('$x_{n}$')
ylabel('$x_{n+1}$')
```

```python
l=3
x=0.4
eps=10**(-2)
X=([])
Y=([])
scatter(x,x,s=15,color='black')
for t in range(0,100):
    new_x1=l*l*x*(1-x)-l**3*x*x*(1-x)**2
    new_x=l*new_x1*(1-new_x1)
    plot([x,x],[x,new_x],color='black')
    plot([x,new_x],[new_x,new_x],color='black')
    #scatter(x,new_x)
    a=x
    x=new_x
if a-x<=eps:
    print('Value converges to',x)
else:
    print('Chaotic')
X=arange(0,1.01,0.01)
Y=l*X*(1-X)
Z=l*l*X*(1-X)-(l**3)*X*X*((1-X)**2)
plot(X,Y,color='grey',linestyle='dotted',linewidth=1.5)
plot(X,Z,color='grey',linestyle='dotted',linewidth=1.5)
plot(X,X,color='grey',linestyle='dashed',linewidth=2)
xlabel('$x_{n}$')
ylabel('$x_{n+2}$')
```

## Fixed points of quadratic map

```python
L=arange(2.8,3.61,0.01)
Lx=([])
X=([])
x=0.5
for l in L:
    for t in range(0,100):
        x=l*x*(1-x)
        Lx=append(Lx,l)
        X=append(X,x)
scatter(Lx,X,s=0.2,color='black')
xlabel('$\lambda$')
ylabel('fixed point')
```

## Henon Map


```python
c=1.4
b=0.3
X=([])
Y=([])
x=0.5;y=0.5
for t in range(0,10000):
    new_x=1-c*x*x+y
    new_y=b*x
    X=append(X,new_x)
    Y=append(Y,new_y)
    x=new_x
    y=new_y
scatter(X,Y,s=1,color='black')
xlabel('x')
ylabel('y')
```
